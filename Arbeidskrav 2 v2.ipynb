# Oppgave 4
### Programmet oppretter en dictionary som som har ulike land og og gir info om hovedstaden i landet og antall innbyggere i mill. i hovedstaden

```python
# a) Opprettelse av dictionary
landsinfo = {
    "Norge": {"hovedstad": "Oslo", "innbyggere": 0.693},
    "Sverige": {"hovedstad": "Stockholm", "innbyggere": 0.975},
    "England": {"hovedstad": "London", "innbyggere": 8.982},
    "Tyskland": {"hovedstad": "Berlin", "innbyggere": 3.769},
    
}

# Program som skriver ut informasjon om et land
land = input("Skriv inn et land: ")
if land in landsinfo:
    hovedstad = landsinfo[land]["hovedstad"]
    innbyggere = landsinfo[land]["innbyggere"]

# Skriver ut resultatet
    
    print(f"{hovedstad} er hovedstaden i {land} og det er {innbyggere} mill. innbyggere i {hovedstad}")
else:
    print(f"Informasjon om {land} finnes ikke i databasen.")

# Program for å legge til nytt land
nytt_land = input("Skriv inn navnet på et nytt land: ")
if nytt_land not in landsinfo:
    hovedstad = input(f"Hva er hovedstaden i {nytt_land}? ")
    innbyggere = float(input(f"Hvor mange millioner innbyggere bor det i {hovedstad}? "))
    landsinfo[nytt_land] = {"hovedstad": hovedstad, "innbyggere": innbyggere}
    print("Landet er lagt til i databasen. Oppdatert dictionary:")
    print(landsinfo)
else:
    print(f"{nytt_land} finnes allerede i databasen.")
```

# Oppgave 5, beregner areal og ytre omkrets av en figur som består av en rettvinklet trekant og en halvsirkel

```python
#importerer nødvendige biblioteker

import math

def beregn_areal_og_omkrets(a, b):
    # Beregn arealet av en figuren
    areal_trekant = 0.5 * a * b
    radius_halvsirkel = b / 2
    areal_halvsirkel = 0.5 * math.pi * (radius_halvsirkel ** 2)
    totalt_areal = areal_trekant + areal_halvsirkel

    # Beregn «ytre» omkrets av en figuren
    hypotenus = math.sqrt(a ** 2 + b ** 2)
    omkrets_halvsirkel = math.pi * radius_halvsirkel
    ytre_omkrets = a + hypotenus + omkrets_halvsirkel

    return totalt_areal, ytre_omkrets

def skriv_ut_resultat():
    try:
        a = float(input("Hva er radius i sirkelen (a): "))
        b = float(input("Hva er lengden i trekanten (b): "))
        
        areal, omkrets = beregn_areal_og_omkrets(a, b)
        print(f"Arealet til figuren er {areal:.2f}.")
        print(f"Den «ytre» omkretsen til figuren er {omkrets:.2f}.")
    except ValueError:
        print("Vennligst oppgi gyldige tall for a og b.")

# Skriver ut resultatet
skriv_ut_resultat()
```
